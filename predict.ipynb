```python
from tensorflow.keras.models import load_model
import cv2
import numpy as np
```

```python
model = load_model("knee_model.keras")
```

```python
IMG_SIZE = 224

def predict_image(path):
    img = cv2.imread(path)
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    img = img / 255.0
    img = np.expand_dims(img, axis=0)

    prediction = model.predict(img)
    class_index = np.argmax(prediction)

    return class_index, prediction
```

```python
result, probs = predict_image(r"C:\Users\shado\OneDrive\Desktop\KNEE\archive (2)\test\3\9113018L.png")

labels = {
    0: "Healthy",
    1: "Doubtful",
    2: "Mild",
    3: "Moderate",
    4: "Severe"
}

print("Prediction:", labels[result])
```
